```python
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
```

```python
from transformers import pipeline
```

```python
Classify = pipeline("ner",model=model, tokenizer= tokenizer)
```

```python
import pdfplumber

# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("token-classification", model="dslim/bert-base-NER")


file_path = "/content/cv3_pdf.pdf"
raw_text = ''
with pdfplumber.open(file_path) as pdf:
  for page in pdf.pages:
        raw_text += page.extract_text()



label_data = pipe(raw_text)
names = ''
for entity in label_data:
  if entity['entity'] == "B-PER":
    name = entity['word']
    names += name
  elif entity['entity'] == "I-PER":
    name = entity['word']
    names += name
    print(names)
print(names.replace('#',''))
```
